#### 2.2.1. Выбор категориальных полей для таблицы Calls
Call Type: анализ типов звонков (входящие/исходящие) поможет понять, какие звонки чаще приводят к положительным результатам.
Call Status: анализ статусов звонков (успешный/пропущенный и т.д.) даст представление о качестве взаимодействия с клиентами.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# Анализ поля 'Call Type'
call_type_counts = calls['Call Type'].value_counts()

# Визуализация
plt.figure(figsize=(10, 6))
plt.pie(call_type_counts, labels=call_type_counts.index, autopct='%1.1f%%', colors=sns.color_palette('coolwarm', len(call_type_counts)))
plt.title('Distribution of Call Types', fontsize=16)
plt.legend(call_type_counts.index, title="Call Type", bbox_to_anchor=(1, 1), loc="upper left", fontsize=12)

plt.show()
```

```python
# Преобразование категориального поля 'Call Status' в строковый тип
calls['Call Status'] = calls['Call Status'].astype(str)

# Анализ поля 'Call Status'
call_status_counts = calls['Call Status'].value_counts().reset_index()
call_status_counts.columns = ['Call Status', 'Number of Calls']

# Визуализация
plt.figure(figsize=(10, 6))
sns.barplot(x='Number of Calls', y='Call Status', data=call_status_counts, palette='viridis', hue='Call Status', dodge=False, legend=False)
plt.title('Distribution of Call Statuses', fontsize=16)
plt.xlabel('Number of Calls', fontsize=14)
plt.ylabel('Call Status', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Добавление меток данных
for index, value in enumerate(call_status_counts['Number of Calls']):
    plt.text(value + 100, index, str(value), va='center', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```

#### 2.2.2. Выбор категориальных полей для таблицы Contacts
Contact Owner Name: анализ распределения количеству контактов по лицам, ответственным за звонок.

```python
# Aгрегируем данные по количеству контактов для каждого владельца; преобразуем поле в строковый тип для корректной сортировки по убыванию
contact_owner_counts = contacts['contact_owner_name'].astype(str).value_counts().sort_values(ascending=False)

# Визуализация данных
plt.figure(figsize=(14, 8))
sns.barplot(x=contact_owner_counts.values, y=contact_owner_counts.index, hue=contact_owner_counts.index, palette='viridis', legend=False)
plt.title('Number of Contacts per Contact Owner', fontsize=16)
plt.xlabel('Number of Contacts', fontsize=14)
plt.ylabel('Contact Owner Name', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Добавление аннотаций для каждого значения на графике
for index, value in enumerate(contact_owner_counts.values):
    plt.text(value + 10, index, str(value), va='center', fontsize=12, color='black')

plt.show()
```

#### 2.2.3. Выбор категориальных полей для таблицы Deals
Quality: aнализ качества лидов помогает выявить, какие категории лидов наиболее ценные для бизнеса, а также определить области для улучшения в процессе привлечения и обработки лидов
Stage: aнализ стадий сделок поможет выявить узкие места в процессе продаж и понять, где теряются потенциальные клиенты.
Source: aнализ источников сделок позволит определить, какие каналы маркетинга наиболее эффективны.

```python
# Преобразование типа данных в строку и исключение значений 'F' и 'Unknown'
deals['Quality'] = deals['Quality'].astype(str)
filtered_deals = deals[~deals['Quality'].isin(['F', 'Unknown'])]

# Подсчет количества сделок для каждой категории 'Quality'
quality_counts = filtered_deals['Quality'].value_counts()

# Построение круговой диаграммы
plt.figure(figsize=(10, 6))
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=sns.color_palette('coolwarm', len(quality_counts)))
plt.title('Distribution of Deals by Quality', fontsize=16)
plt.axis('equal')
plt.legend(quality_counts.index, title="Quality", bbox_to_anchor=(1, 1), loc="upper left", fontsize=12)
plt.show()
```

#### ВЫВОД: 
Наиболее значительная доля (39.5%) приходится на категорию "E - Non Qualified", что указывает на высокую долю неподходящих лидов.

К категории "D - Non Target" относится 32.4% сделок, а категории "C - Low" и "B - Medium" занимают 17.8% и 8% соответственно.

Высококачественные лиды "A - High" составляют лишь 2.2% от общего числа сделок, что указывает на необходимость улучшения процесса привлечения и квалификации лидов.

```python
# Анализ поля 'Stage'
stage_counts = deals['Stage'].value_counts()

# Исключение категории 'Unknown'
stage_counts = stage_counts[stage_counts.index != 'Unknown']

# Сортировка данных по убыванию
stage_counts = stage_counts.sort_values(ascending=True)

# Визуализация с помощью горизонтальной столбчатой диаграммы
plt.figure(figsize=(12, 8))
bars = plt.barh(stage_counts.index, stage_counts.values, color=sns.color_palette('Greens_d', len(stage_counts)))

# Добавление меток данных
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f' {bar.get_width()}', va='center', fontsize=12, color='black')

plt.title('Deal Stage Distribution', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Stage', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Анализ поля 'Source', не отображаем данные со значением Test, поскольку это тестовые данные, а не реальные
source_counts = spend.loc[spend['Source'] != 'Test', 'Source'].value_counts()

# Сортировка данных по возрастанию
source_counts = source_counts.sort_values(ascending=True)

# Визуализация с помощью горизонтальной накопительной столбчатой диаграммы
plt.figure(figsize=(12, 8))
bars = plt.barh(source_counts.index, source_counts.values, color=sns.color_palette('Oranges_d', len(source_counts)))

# Добавление меток данных
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f' {bar.get_width()}', va='center', fontsize=12, color='black')

plt.title('Marketing Source Distribution', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Source', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

#### 2.2.4. Выбор категориальных полей для таблицы Spend
Проанализируем данные по суммам Spend по каждому каналу Source и построим график, который показывает суммарные значения Spend для каждого источника

```python
# Агрегирование данных по суммам 'Spend' для каждого 'Source', исключая 'Test'
total_spend_per_source = spend.loc[spend['Source'] != 'Test'].groupby('Source', observed=True)['Spend'].sum().reset_index()

# Сортировка данных по возрастанию суммы расходов
total_spend_per_source = total_spend_per_source.sort_values(by='Spend', ascending=True)

# Визуализация с помощью горизонтальной накопительной столбчатой диаграммы
plt.figure(figsize=(14, 10))
bars = plt.barh(total_spend_per_source['Source'], total_spend_per_source['Spend'], color=sns.color_palette('Oranges_d', len(total_spend_per_source)))

# Добавление меток данных
for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():,.2f}€', va='center', fontsize=12, color='black')

plt.title('Total Spend by Source', fontsize=16)
plt.xlabel('Total Spend', fontsize=14)
plt.ylabel('Source', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```
